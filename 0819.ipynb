```python
import numpy as np
import matplotlib.pyplot as plt

def mls(x, y):
    mx = np.mean(x)
    my = np.mean(y)
    # 최소 제곱법의 분모 코드
    divisor = sum([(i - mx)**2 for i in x])
    # 최소 제곱법의 분자 코드
    r = zip(x, y)
    dividend = sum([(x-mx)*(y-my) for x, y in r])

    a = dividend / divisor
    b = my - (mx*a)
    return (a, b)


# y = a * x + b
x = np.array([2, 4, 6, 8])
y = np.array([81, 93, 91, 97])

aa, bb = mls(x, y)
print('기울기 a는', aa)
print('절편 b는', bb)

plt.scatter(x, y)
plt.plot([0, 10], [aa*0+bb, aa*10+bb], 'r')
```

```python
# 경사하강법을 왜 하는가?
# 기울기 a와 절편 b를 구하기 위해서 사용한다
# 최소 제곱법 공식으로 이미 구할 수 있는 것 아닌가?
# 최소 제곱법은 1차 방정식에서만 성립된다
# y = ax + b >> 최소제곱법, 경사하강법
# y = ax + by + cz + d >> 경사하강법

import numpy as np
import matplotlib.pyplot as plt

# 기울기: gradient >> 경사
# y = a * x + 79
# a = 2.3 일 때 오차값이 제일 작다
gradient = [1.8, 1.9, 2.0, 2.1, 2.2, 2.3, 2.4, 2.5, 2.6, 2.7, 2.8]
intercept = 79
x = np.array([2, 4, 6, 8])
y = np.array([81, 93, 91, 97])

lose_list = []

for index in range(11):
    predict_list = []
    for i in range(4) :
        #predict = gradient[i] * x[i] + intercept
        predict = gradient[index] * x[i] + intercept
        predict_list.append(predict)
        
    #print( y - predict_list)
    #print((y - predict_list) ** 2)
    #print( sum((y - predict_list) ** 2))
    #print(gradient[index],  sum((y - predict_list) ** 2) / 4)
    lose_value = sum((y - predict_list) ** 2) / 4
    lose_list.append(lose_value)
# 기울기    오차값 
# 1.8       15.79999999

plt.xlabel('gradient(a)')
plt.ylabel('lose value')
plt.scatter(gradient, lose_list)
plt.plot(gradient, lose_list, 'r')
```
